```python
import numpy as np
import matplotlib.pyplot as plt

months = np.array(["Jan", "Feb", "Mar", "Apr", "May", "Jun",
                   "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"])
temps = np.array([25, 27, 30, 33, 35, 38, 37, 36, 34, 31, 28, 26])

avg_temp = np.mean(temps)
hottest_month = months[np.argmax(temps)]
coldest_month = months[np.argmin(temps)]
norm_temps = (temps - np.min(temps)) / (np.max(temps) - np.min(temps))
above_avg_months = months[temps > avg_temp]
temps_f = temps * 9/5 + 32

print("Average Temperature:", avg_temp)
print("Hottest Month:", hottest_month)
print("Coldest Month:", coldest_month)
print("Normalized Temperatures:", norm_temps)
print("Months Above Average:", above_avg_months)calca
print("Temperatures in Fahrenheit:", temps_f)

plt.plot(months, temps, marker='o')
plt.xlabel("Month")
plt.ylabel("Temperature (°C)")
plt.title("Monthly Temperature Trend")
plt.show()
```

```python
import pandas as pd

data = {
    "OrderID": [101, 102, 103, 104, 105],
    "Product": ["Laptop", "Phone", "Tablet", "Monitor", "Headphones"],
    "Quantity": [2, 5, 3, 1, 4],
    "Price": [50000, 15000, 20000, 12000, 3000]
}
df = pd.DataFrame(data)
df["TotalSale"] = df["Quantity"] * df["Price"]
filtered = df[df["TotalSale"] > 1000]
most_sold = df.groupby("Product")["Quantity"].sum().idxmax()
df["Discount"] = df["TotalSale"].apply(lambda x: 0.1*x if x > 20000 else 0.05*x)

print(df)
print("Filtered Orders:\n", filtered)
print("Most Sold Product:", most_sold)

df.to_excel("sales_summary.xlsx", index=False)
```
